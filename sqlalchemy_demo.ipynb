### This is a very simple demo using sqlalchemy to engage a database and create a dataframe from the query results.

```
from sqlalchemy import create_engine
import os
import configparser
import pandas as pd
import numpy as np
```

```
home = os.path.expanduser('~')
cf = home+'/.jira.ini'

config = configparser.ConfigParser()
config.read(cf)

u = config['PGSQL']['username']
p = config['PGSQL']['password']
url = config['PGSQL']['url']
```

```
prod_uri = f'postgresql://{u}:{p}@{url}'
sql = "SELECT * from polarion.workitem where c_type = 'requirement'"
```

```
engine = create_engine(prod_uri, client_encoding='utf8')
```

```
with engine.connect() as con:
    con.execute('SET search_path TO polarion')
    rs = con.execute(sql)
    df = pd.DataFrame(rs.fetchall())
    
```
